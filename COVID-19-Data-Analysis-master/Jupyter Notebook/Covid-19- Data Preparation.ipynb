### Data Transformation

```python
# convert observation date into a date
covid_df["ObservationDate"]=pd.to_datetime(covid_df["ObservationDate"])
```

```python
covid_df['Country/Region'].replace(
    to_replace=["('St. Martin',)"],
    value='St. Martin',
    inplace=True
)
covid_df1 = covid_df['Country/Region'].str.contains('St. Martin') 
covid_df1
```

```python
covid_df["Country/Region"].explode().value_counts()
```

```python
# Channel Islands +  UK + North Ireland to United Kingdom
covid_df["Country/Region"] = covid_df["Country/Region"].replace({"UK": "United Kingdom",
                                                   "Channel Islands": "United Kingdom",
                                                    "North Ireland": "United Kingdom"}
                                                   )
```

### Include Latitude and Longitude

```python
# adding lat and long
centroids = pd.read_csv("world_country_and_usa_states_latitude_and_longitude_values.csv")
covid_new_df = pd.read_csv('C_Corona.csv')
centroids.columns
```

```python
centroids.drop(['country_code','usa_state_code',
       'usa_state_latitude', 'usa_state_longitude', 'usa_state'], axis =1, inplace=True)
centroids.head()
```

```python
centroids["country"].value_counts()
```

```python
centroids["country"] = centroids["country"].replace({"Congo [Republic]": "Congo"})
#WARNING:root:Netherlands Antilles not found in regex
#WARNING:root:Congo [Republic] not found in regex
```

```python
centroids = centroids[centroids["country"] != "Netherlands Antilles"]  
                      
```

```python
countr3= centroids["country"].to_list()
print(countr3)
```

```python
standard_names3 = coco.convert(names=countr3, to='name_short')
print(standard_names3)
```

```python
data_tuples3 = list(zip(countr3, standard_names3))
dic3 = dict(data_tuples3)
dic3
```

```python
centroids.loc[:,"country"].replace(to_replace = dic3.keys(), value = dic3.values() , inplace=True)
centroids.head()
```

```python
centroids =centroids.drop_duplicates(subset=['country'], keep='last')
centroids.shape
```

<b>Joining two dataframes to have the Continent, Logitude and Latitude information in the dataframe</b>

```python
#merging the columns Lon and Lat
#covid_complete_df=pd.merge(covid_new_df,centroids[["Country/Region", "Longitude"]], how="left", left_on = 'Country/Region', right_on = 'sovereignt').drop('sovereignt')
covid_complete_df = pd.merge(left=covid_new_df, right=centroids, how='left', left_on='Country/Region', right_on='country').set_index('ObservationDate')
del covid_complete_df["country"]
#del covid_complete_df["Last Update"]
#del covid_complete_df["Unnamed: 0"]
```
